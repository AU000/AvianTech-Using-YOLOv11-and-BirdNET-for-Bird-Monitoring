```python
# Cell 1: Complete GPU Detection and Verification

import torch
import sys

print("="*80)
print("GPU DETECTION AND CONFIGURATION CHECK")
print("="*80)

# Check PyTorch version
print(f"\n1. PyTorch Version: {torch.__version__}")
print(f"   Python Version: {sys.version}")

# Check CUDA availability
print(f"\n2. CUDA Available: {torch.cuda.is_available()}")

if torch.cuda.is_available():
    print("   ✅ GPU is available for training!")
    
    # Get CUDA version
    print(f"\n3. CUDA Version: {torch.version.cuda}")
    
    # Get number of GPUs
    gpu_count = torch.cuda.device_count()
    print(f"\n4. Number of GPUs: {gpu_count}")
    
    # Get GPU details for each device
    print(f"\n5. GPU Details:")
    for i in range(gpu_count):
        print(f"\n   GPU {i}:")
        print(f"   - Name: {torch.cuda.get_device_name(i)}")
        print(f"   - Memory Allocated: {torch.cuda.memory_allocated(i) / 1024**3:.2f} GB")
        print(f"   - Memory Cached: {torch.cuda.memory_reserved(i) / 1024**3:.2f} GB")
        print(f"   - Total Memory: {torch.cuda.get_device_properties(i).total_memory / 1024**3:.2f} GB")
    
    # Set default GPU
    print(f"\n6. Current GPU Device: {torch.cuda.current_device()}")
    print(f"   Device Name: {torch.cuda.get_device_name(torch.cuda.current_device())}")
    
    # Test GPU with a simple operation
    print(f"\n7. GPU Functionality Test:")
    try:
        x = torch.rand(5, 3).cuda()
        print("   ✅ GPU tensor creation successful!")
        print(f"   Test tensor device: {x.device}")
        del x
        torch.cuda.empty_cache()
    except Exception as e:
        print(f"   ❌ GPU test failed: {e}")
    
else:
    print("   ❌ GPU is NOT available")
    print("   ⚠️  Training will use CPU (much slower)")
    print("\n   Possible reasons:")
    print("   - No NVIDIA GPU installed")
    print("   - CUDA drivers not installed")
    print("   - PyTorch installed without CUDA support")
    print("\n   To install PyTorch with CUDA support, visit:")
    print("   https://pytorch.org/get-started/locally/")

print("\n" + "="*80)
```

```python
# Cell 3: NVIDIA GPU Check (if on Linux/Windows)

# For systems with nvidia-smi
import subprocess
import sys

print("="*80)
print("NVIDIA GPU SYSTEM CHECK")
print("="*80)

try:
    result = subprocess.run(['nvidia-smi'], 
                          stdout=subprocess.PIPE, 
                          stderr=subprocess.PIPE, 
                          text=True)
    
    if result.returncode == 0:
        print("\n✅ NVIDIA GPU detected by system:\n")
        print(result.stdout)
    else:
        print("\n❌ nvidia-smi command failed")
        print("NVIDIA drivers may not be installed")
        
except FileNotFoundError:
    print("\n⚠️  nvidia-smi not found")
    print("This means either:")
    print("  - No NVIDIA GPU is present")
    print("  - NVIDIA drivers are not installed")
    print("  - Running on Mac (Apple Silicon or Intel)")
    
except Exception as e:
    print(f"\n❌ Error checking GPU: {e}")
```

```python
# Cell 4: Detailed GPU Memory Check

import torch

if torch.cuda.is_available():
    print("="*80)
    print("GPU MEMORY ANALYSIS")
    print("="*80)
    
    for i in range(torch.cuda.device_count()):
        print(f"\nGPU {i}: {torch.cuda.get_device_name(i)}")
        print("-" * 60)
        
        # Get memory info
        total_memory = torch.cuda.get_device_properties(i).total_memory / 1024**3
        allocated_memory = torch.cuda.memory_allocated(i) / 1024**3
        reserved_memory = torch.cuda.memory_reserved(i) / 1024**3
        free_memory = total_memory - reserved_memory
        
        print(f"Total Memory:     {total_memory:.2f} GB")
        print(f"Allocated Memory: {allocated_memory:.2f} GB")
        print(f"Reserved Memory:  {reserved_memory:.2f} GB")
        print(f"Free Memory:      {free_memory:.2f} GB")
        
        # Recommend batch size based on available memory
        print(f"\n💡 Recommended Settings for This GPU:")
        if total_memory >= 24:
            print("   - Batch Size: 32-64")
            print("   - Model: YOLOv5s, YOLOv5m, or YOLOv5l")
        elif total_memory >= 12:
            print("   - Batch Size: 16-32")
            print("   - Model: YOLOv5s or YOLOv5m")
        elif total_memory >= 6:
            print("   - Batch Size: 8-16")
            print("   - Model: YOLOv5n or YOLOv5s")
        elif total_memory >= 4:
            print("   - Batch Size: 4-8")
            print("   - Model: YOLOv5n")
        else:
            print("   - Batch Size: 2-4")
            print("   - Model: YOLOv5n")
            print("   ⚠️  Limited GPU memory - consider using CPU or cloud GPU")
else:
    print("\n❌ No GPU available for memory analysis")
```

```python
# Cell 5: Quick GPU Benchmark

import torch
import time

print("="*80)
print("GPU PERFORMANCE BENCHMARK")
print("="*80)

if torch.cuda.is_available():
    # Warm up
    x = torch.randn(1000, 1000).cuda()
    y = torch.randn(1000, 1000).cuda()
    _ = torch.matmul(x, y)
    torch.cuda.synchronize()
    
    # Benchmark
    print("\nRunning benchmark... (matrix multiplication)")
    
    start = time.time()
    for _ in range(100):
        z = torch.matmul(x, y)
    torch.cuda.synchronize()
    gpu_time = time.time() - start
    
    print(f"✅ GPU Time: {gpu_time:.4f} seconds")
    
    # Compare with CPU
    x_cpu = x.cpu()
    y_cpu = y.cpu()
    
    start = time.time()
    for _ in range(100):
        z_cpu = torch.matmul(x_cpu, y_cpu)
    cpu_time = time.time() - start
    
    print(f"⚠️  CPU Time: {cpu_time:.4f} seconds")
    print(f"\n🚀 GPU is {cpu_time/gpu_time:.2f}x faster than CPU")
    
    # Clean up
    del x, y, x_cpu, y_cpu
    torch.cuda.empty_cache()
    
else:
    print("\n❌ No GPU available for benchmark")
    print("Training will run on CPU (significantly slower)")
```
