```python
import json
import os
from pathlib import Path

# Load COCO annotations
coco_path = r"C:\Users\Christian\AvianTech\Birds_Cv.coco\train\_annotations.coco.json"
images_path = r"C:\Users\Christian\AvianTech\Birds_Cv.coco\train"
output_labels = r"C:\Users\Christian\AvianTech\Birds_Cv.coco\train\labels"

os.makedirs(output_labels, exist_ok=True)

with open(coco_path) as f:
    coco = json.load(f)

# Get image dimensions
image_info = {img["id"]: img for img in coco["images"]}

# Get categories
categories = {cat["id"]: cat["name"] for cat in coco["categories"]}
print("Bird species found:", categories)

# Convert annotations
for ann in coco["annotations"]:
    img = image_info[ann["image_id"]]
    img_w, img_h = float(img["width"]), float(img["height"])
    
    x, y, w, h = [float(v) for v in ann["bbox"]]  # ← fix here
    x_center = (x + w / 2) / img_w
    y_center = (y + h / 2) / img_h
    w_norm = w / img_w
    h_norm = h / img_h
    
    cat_id = ann["category_id"] - 1  # YOLO classes start at 0
    
    label_file = os.path.join(output_labels, Path(img["file_name"]).stem + ".txt")
    with open(label_file, "a") as f:
        f.write(f"{cat_id} {x_center:.6f} {y_center:.6f} {w_norm:.6f} {h_norm:.6f}\n")

print(f"✅ Converted {len(coco['annotations'])} annotations!")
print(f"✅ Labels saved to: {output_labels}")
```

```python
import shutil
import random
import os
from pathlib import Path

# Paths
src_images = r"C:\Users\Christian\AvianTech\Birds_Cv.coco\train"
src_labels = r"C:\Users\Christian\AvianTech\Birds_Cv.coco\train\labels"

base = r"C:\Users\Christian\AvianTech\data\datasets"

# Get all images only
images = [f for f in os.listdir(src_images) if f.endswith((".jpg", ".png", ".jpeg"))]
random.seed(42)
random.shuffle(images)

# Calculate splits
total = len(images)
train_end = int(total * 0.7)
val_end = int(total * 0.9)

split_data = {
    "train": images[:train_end],
    "val": images[train_end:val_end],
    "test": images[val_end:]
}

# Copy files
for split, files in split_data.items():
    img_dir = os.path.join(base, split, "images")
    lbl_dir = os.path.join(base, split, "labels")
    os.makedirs(img_dir, exist_ok=True)
    os.makedirs(lbl_dir, exist_ok=True)
    
    for fname in files:
        # Copy image
        shutil.copy(os.path.join(src_images, fname), os.path.join(img_dir, fname))
        # Copy label
        label_fname = Path(fname).stem + ".txt"
        label_src = os.path.join(src_labels, label_fname)
        if os.path.exists(label_src):
            shutil.copy(label_src, os.path.join(lbl_dir, label_fname))

    print(f"✅ {split}: {len(files)} images")

print(f"\n📊 Total images: {total}")
print("🐦 Dataset split complete!")
```
