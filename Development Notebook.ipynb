```python
# Verify NVIDIA GPU is available and configured properly

import sys
import torch
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from pathlib import Path
import os

print("="*60)
print("GPU & ENVIRONMENT VERIFICATION")
print("="*60)
print(f"Python version: {sys.version}")
print(f"PyTorch version: {torch.__version__}")

# Check CUDA availability
print(f"\n🎮 CUDA available: {torch.cuda.is_available()}")

if torch.cuda.is_available():
    print(f"✅ GPU Device: {torch.cuda.get_device_name(0)}")
    print(f"✅ GPU Count: {torch.cuda.device_count()}")
    print(f"✅ Current GPU: {torch.cuda.current_device()}")
    print(f"✅ CUDA Version: {torch.version.cuda}")
    
    # Check GPU memory
    gpu_memory = torch.cuda.get_device_properties(0).total_memory / 1024**3
    print(f"✅ GPU Memory: {gpu_memory:.2f} GB")
    
    # Set GPU as default device
    torch.cuda.set_device(0)
    print("\n✅ GPU is ready for training!")
else:
    print("\n⚠️ WARNING: CUDA not available. Training will use CPU (slower)")
    print("Please install CUDA-enabled PyTorch:")
    print("pip install torch torchvision --index-url https://download.pytorch.org/whl/cu118")

print(f"\nOpenCV version: {cv2.__version__}")
print(f"NumPy version: {np.__version__}")
print("="*60)
```
